```python
# Who is the political leader of the 2nd highest populated country
df['political_leader'][df['population'] == df['population'].nlargest(2).iloc[1]]
```

```python
# How many countries have Republic in their full name
count = 0
def extraction(txt):
    global count
    if 'republic' in txt.lower():
        count += 1
    return txt
df['country_long'] = df['country_long'].apply(extraction)
print(count)
```
