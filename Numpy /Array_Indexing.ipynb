```python
import numpy as np
```

```python
# ----------------------Numpy Indexing and Slicing of vectors----------------------
```

```python
arr = np.arange(11, 21)
arr
```

```python
arr[5]
```

```python
arr[1:5]
```

```python
arr[3::2]
```

```python
# ----------------------Numpy Indexing and Slicing of Matrix----------------------
```

```python
arr = np.arange(1, 31).reshape(6, 5)
arr
```

```python
arr[3], arr[5], arr[1]
```

```python
arr[0,4], arr[5,4]
```

```python
arr[0:2, 1:3]
```

```python
arr[3:5, 3:4]
```

```python
arr[3:, 2:]
```

```python
arr[:, 3]
```
